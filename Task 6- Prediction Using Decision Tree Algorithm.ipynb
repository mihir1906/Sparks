# Task 6: Prediction Using Decision Tree Algorithm

- Create the Decision Tree classifier and visualize it graphically.

```python
#Importing Libraries
import pandas as pd
from sklearn.datasets import load_iris
```

```python
#Loading the dataset
iris = load_iris()
iris
```

```python
#Forming the iris dataframe
df_iris = pd.DataFrame(iris.data , columns = iris.feature_names)
X = df_iris #Features
X.head()

```

```python
Y = iris.target #Target
Y
```

```python
#Spliting the data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(iris.data, iris.target)
X_train, X_test, Y_train, Y_test
```

```python
#Training the decision tree algorithm
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,Y_train)
```

```python
#Testing the Algorithm
prediction = dt.predict(X_test) 
comp_pred = pd.DataFrame({'Prediction': prediction, 'Actual Value': Y_test})
species = {0:'Iris-setosa',1:'Iris-versicolor',2:'Iris-virginica'}
comp_pred['Species'] = comp_pred['Prediction'].map(species)
comp_pred
```

```python
#Making Predictions
pred = [[ 5.7 , 3.8 ,1.2 ,0.1]]
final = dt.predict(pred)
output = pd.DataFrame({'Prediction':final})
output['Species'] = output['Prediction'].map(species)
output
```

```python
#Accuracy
dt.score(iris.data, iris.target)
```
