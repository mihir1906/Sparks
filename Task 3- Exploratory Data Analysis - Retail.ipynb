```python
#Importing  required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading the data
data = pd.read_csv('C:\\Users\\mihir\\Downloads\\SampleSuperstore.csv')
```

```python
#Checking for the null values
data.isnull() == False
```

```python
#Counting the number of unique values in each column
data.nunique(axis = 0)
```

```python
#Checking for duplicate values
data[data.duplicated(keep = 'first')]
```

```python
#Clearing all duplicate values
data.drop_duplicates(inplace=True)
data.shape
```

```python
#sorting the regions by their profits in descending order
sales = data.groupby('Region').sum()['Sales'].sort_values(ascending=False).reset_index()
```

```python
plt.figure(figsize = (10,10))
plt.title('Regional Sales')
sns.barplot(x='Region',y ='Sales', data = sales.head(7))
plt.xticks(rotation = 45)
plt.show()
```

```python
#Ploting the distplot for discount
sns.distplot(data['Discount'], bins = 10)
```

```python
corr = data.corr()
sns.heatmap(corr, annot = True , cmap = 'PuRd', square = True, cbar = True)
```

```python
sns.scatterplot(x='Quantity',y='Sales', data = data)
plt.title('Quantity vs. Sales')
```

```python
plt.figure(figsize =(10,10))
sns.countplot(x= data['City'][:20])
plt.xticks(rotation = 45)
plt.show()
```

```python
plt.figure(figsize=(20,15))
sns.pairplot(data, hue='Sub-Category')
```

```python
sns.boxplot(y = 'Profit',data = data, width = 5)
```

```python
sns.lineplot(x='Discount', y = 'Profit', data = data)
```

```python
sns.lineplot(x='Sales', y = 'Profit', data = data)
```

```python
con = data.groupby('Region')[['Sales','Profit']].sum().sort_values(by = 'Sales', ascending = True)
con[:].plot.bar(color = ['Purple', 'Green'], figsize = (10,10))
plt.title('Sales and Profit across Region')
plt.show()
```

 - Thus we conclude that the profits decreases as the sales and discounts increases
